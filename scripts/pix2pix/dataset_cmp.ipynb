```python
import pandas as pd

df_real = pd.read_csv('../../assets/result/images_real/result.csv')
df_real.head(10)
```

```python
df_fake = pd.read_csv('../../assets/result/images_pred/result.csv')
df_fake.head(10)
```

```python
columns = ['area', 'circumstance', 'circularity', 'finger_num', 'velocity', 'circle_radius']  # 替换为你要比较的列名

diff_dict = {}
for column in columns:
    diff_dict[column] = ((df_fake[column] - df_real[column]) / df_real[column])

diff_df = pd.DataFrame(diff_dict)
diff_df
```
