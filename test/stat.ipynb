```python
import pandas as pd
import matplotlib.pyplot as plt

label = 'finger_num'
df = pd.read_csv(f'../assets/test-case/{label}.csv')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create more detailed bins for values under 10%, and fewer bins for higher values
bins = [0, 3, 6, 10, 15, 20, 50, 100]
labels = ['0-3%', '3-6%', '6-10%', '10-15%', '15-20%', '20-50%', '50-100%']

# Count the frequency in each interval
hist_data = pd.cut(df['percentage_error'], bins=bins, labels=labels).value_counts().sort_index()

# Create bar chart
plt.figure(figsize=(12, 6))
hist_data.plot(kind='bar')

# Set chart title and axis labels
plt.title(f'Percentage Error Distribution(label = {label})')
plt.xlabel('Percentage Intervals')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display values on top of each bar
for i, v in enumerate(hist_data):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the chart
plt.savefig(f'../assets/test-case/fig-{label}')
plt.show()
```
